```python
import numpy as np
a = np.array([1, 2, 3, 7, 8, 24])
b = np.zeros((5, 6))
c = np.ones((5, 5))
d = np.arange(2, 8)
print(a, '\n', b, '\n', c, '\n', d)
```

```python
print(a.shape, a.dtype, type(a), '\n',
     b.shape, b.dtype, type(b), '\n',
     c.shape, c.dtype, type(c), '\n',
     d.shape, d.dtype, type(d), '\n',)
```

```python
print(a[1], a[[0, 2]], a[0:2],'\n', b[0][1], b[[1, 3], [3, 3]],'\n', b[2:5, 2:5])
```

```python
z = a*d
print(z)
x = np.mean(z)/(z.max()+z.min())
print(x)
```
